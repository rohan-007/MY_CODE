```python
# IMPORTING MODULES AND DATASET FROM SKLEARN
import numpy as np
import pandas as pd
from sklearn import datasets

```

```python
# ANALYSING THE CANCER DATASET
cancer=datasets.load_breast_cancer()
print(type(cancer))
print(cancer.DESCR)
```

```python
# FETCHING LIST OF FEATURES AND TARGET VARIALBES
flist=cancer.feature_names
tlist=cancer.target_names
print(flist,tlist,sep='\n')
print(len(flist),len(tlist),sep=';')
```

```python
# FEATCHING FEATURES AND TARGET VARIABLES IN ARRAY FORMAT ,SO THAT WE CAN PASS IT TO TRAIN_TEST_SPLIT
x=cancer.data
y=cancer.target
print(x,y,sep='\n')
```

```python
# CONVERTING BUNCH DATATYPE TO PANDAS DATAFRAME
dataframe=pd.DataFrame(x)
df=dataframe
dataframe.head(6)
```

```python
# RENAMING THE COLUMNS OF CREATED DATAFRAME
lst1=list(range(len(flist)))
flist=cancer.feature_names
map_lst1_flist=dict(zip(lst1,flist))

df.rename(columns=map_lst1_flist)
```
