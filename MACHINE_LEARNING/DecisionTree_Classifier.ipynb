# NO-OVERFITTING

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris()
x=iris.data[:,2:]
y=iris.target
flist=iris.feature_names
print(flist)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
tree_clf=DecisionTreeClassifier(max_depth=2,random_state=1)
tree_clf.fit(x_train,y_train)
```

```python
from sklearn import tree
fig=plt.figure(figsize=(6,6))
tree.plot_tree(tree_clf,feature_names=iris.feature_names[2:])    
```

```python
tree_clf.predict([[5,0.9]])
```

```python
predict=tree_clf.predict(x_test)
predict_proba=tree_clf.predict_proba(x_test)
print(tree_clf.score(x_test,y_test))
```

```python
print(y_test)
```

```python
print(predict)
```

```python
import numpy as np
x_range=np.arange(x[:,0].min()-1,x[:,0].max()+1,0.01)
y_range=np.arange(x[:,1].min()-1,x[:,1].max()+1,0.01)
xx,yy=np.meshgrid(x_range,y_range)
print(xx.shape,yy.shape,sep='\n')
print('Total co-ordinates:',790*440)
```

```python
input_array=np.array([xx.ravel(),yy.ravel()]).T
labels=tree_clf.predict(input_array)
```

```python
import matplotlib.pyplot as plt
plt.contourf(xx,yy,labels.reshape(xx.shape),alpha=0.3)
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
plt.xlabel('Petal_Length --->')
plt.ylabel('Petal_width --->')
plt.title('Decision_Boundary')
```
