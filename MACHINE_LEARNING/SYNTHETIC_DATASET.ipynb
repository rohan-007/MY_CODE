```python
# CREATION OF SIMPLE LINEAR REGRESSION { UNI-VARIATE}
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
plt.figure()
plt.title('SYNTHETICALLY CREATED REGRESSION PROBLEM')
x,y=make_regression(n_samples=100,n_features=1,n_informative=1,noise = 10,random_state=10)
plt.scatter(x,y,s=30,marker='o',color='red')
plt.show()
print(x.shape,y.shape,sep='\n')
```

```python
# CREATION OF NON-LINEAR REGRESSION { COMPLEX - REGRESSION }
from sklearn.datasets import make_friedman1
plt.figure()
plt.title('SYNTHETICALLY PRODUCE COMPLEX - REGRESSION PROBLEM')
x_com,y_com=make_friedman1(n_samples=100,n_features=7,noise=0,random_state=2)
plt.scatter(x_com[:,2],y_com,marker='o',s=30,color='red')
plt.show()
print(x_com.shape,y_com.shape,sep='\n')
```

```python
# CREATING BINARY-CLASS CLASSIFICATION DATASET { 2-Features} and { linearly-seperable}
from sklearn.datasets import make_classification
plt.figure()
plt.title('SYNTHETICALLY PRODUCED BINATY-CLASS CLASSIFICATION')
x_b,y_b=make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,n_classes=2,random_state=1,n_clusters_per_class=1,flip_y = 0.1)
plt.scatter(x_b[:,0],x_b[:,1],c=y_b,s=30)
```

```python
# CREATING BINARY-CLASS CLASSIFICATION DATASET { 2-FEATURES} AND { NON-LINAERLR -SEPERABLE}
from sklearn.datasets import make_blobs
plt.figure()
plt.title('CREATING NON-LINEARLY SEPERABLE BINARY-CLASS CLASSIFICATION')
x_bc,y_bc=make_blobs(n_samples=100,n_features=2,random_state=1,centers=8)
y_bc=y_bc%2 # used to convert multi-class to binary-class
plt.scatter(x_bc[:,0],x_bc[:,1],c=y_bc)
plt.show()
```
