```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
plt.figure()
plt.title('SIMPLE-LINEAR-REGRESSION')
x,y=make_regression(n_samples=100,n_features=1,n_informative=1,noise=15,random_state=3)
plt.scatter(x,y,color='red',marker='o',s=30)


```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
knn=KNeighborsRegressor(n_neighbors=10)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
knn.fit(x_train,y_train)
predict=knn.predict(x_test)
print(knn.score(x_test,y_test),predict,sep='\n')

```

```python
plt.scatter(x_train,y_train,color='red' )
plt.scatter(x_test,predict,marker='^',s=90)
```

```python
import numpy as np
x_new=np.linspace(-3,2,100).reshape(100,1)
predict_new=knn.predict(x_new)
plt.plot(x_new,predict_new,color='blue')
plt.scatter(x_train,y_train,color='red' )
plt.scatter(x_test,predict,marker='^',s=90)
```
