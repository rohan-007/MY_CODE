```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from math import *

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import jaccard_similarity_score,accuracy_score,log_loss,classification_report,confusion_matrix
```

# TRAIN AND TEST

```python
x=np.asanyarray(data[['PetalLengthCm','PetalWidthCm']])
y=np.asanyarray(data['Species'])
scaler=MinMaxScaler()
x_processed=scaler.fit_transform(x)
print(x,x_processed,sep='\n')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_processed, y, test_size=0.2, random_state=3)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
predict=knn.predict(x_test)
acc=knn.score(x_test,y_test)
print(acc)
```
